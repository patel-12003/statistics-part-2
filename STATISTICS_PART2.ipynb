```python
#  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import math
from scipy import stats

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):

    z_statistic = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))


    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))


    print(f"Z-statistic: {z_statistic}")
    print(f"P-value: {p_value}")

    if p_value < alpha:
        print(f"Since the p-value ({p_value}) is less than alpha ({alpha}), we reject the null hypothesis.")
        print("Conclusion: The sample mean is significantly different from the population mean.")
    else:
        print(f"Since the p-value ({p_value}) is greater than alpha ({alpha}), we fail to reject the null hypothesis.")
        print("Conclusion: The sample mean is not significantly different from the population mean.")


sample_mean = 105
population_mean = 100
population_std = 15
sample_size = 30


z_test(sample_mean, population_mean, population_std, sample_size)

```

```python
# Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy import stats
import math

def z_test_one_sample(sample_data, population_mean, population_std, alpha=0.05):

    sample_mean = np.mean(sample_data)


    n = len(sample_data)


    z_statistic = (sample_mean - population_mean) / (population_std / math.sqrt(n))


    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))


    print(f"Sample Mean: {sample_mean}")
    print(f"Z-statistic: {z_statistic}")
    print(f"P-value: {p_value}")


    if p_value < alpha:
        print(f"Since the p-value ({p_value}) is less than alpha ({alpha}), we reject the null hypothesis.")
        print("Conclusion: The sample mean is significantly different from the population mean.")
    else:
        print(f"Since the p-value ({p_value}) is greater than alpha ({alpha}), we fail to reject the null hypothesis.")
        print("Conclusion: The sample mean is not significantly different from the population mean.")


np.random.seed(0)


sample_data = np.random.normal(loc=105, scale=10, size=30)


population_mean = 100
population_std = 15


z_test_one_sample(sample_data, population_mean, population_std)
```

```python
# Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_type_errors(sample_mean, population_mean, population_std, sample_size, alpha=0.05):

    z_critical = stats.norm.ppf(1 - alpha / 2)
    critical_value = population_mean + z_critical * (population_std / np.sqrt(sample_size))



    true_mean = 110
    z_beta = (critical_value - true_mean) / (population_std / np.sqrt(sample_size))
    beta = stats.norm.cdf(z_beta)


    x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)


    y_null = stats.norm.pdf(x, population_mean, population_std / np.sqrt(sample_size))
    plt.plot(x, y_null, label="Null Hypothesis Distribution", color='blue')


    y_alt = stats.norm.pdf(x, true_mean, population_std / np.sqrt(sample_size))
    plt.plot(x, y_alt, label="Alternative Hypothesis Distribution (True Mean)", color='orange')


    plt.fill_between(x, y_null, where=(x > critical_value), color='red', alpha=0.5, label="Type I Error Region (Reject Null when True)")


    plt.fill_between(x, y_alt, where=(x < critical_value), color='green', alpha=0.5, label="Type II Error Region (Fail to Reject when False)")


    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()


    print(f"Critical value for rejection region: {critical_value:.2f}")
    print(f"Type I error (alpha) = {alpha}")
    print(f"Type II error (beta) = {beta:.4f}")


np.random.seed(0)


population_mean = 100
population_std = 15
sample_size = 30
sample_mean = 105


visualize_type_errors(sample_mean, population_mean, population_std, sample_size)
```

```python
# Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
from scipy import stats

def independent_t_test(group1, group2, alpha=0.05):

    t_stat, p_value = stats.ttest_ind(group1, group2)


    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")


    if p_value < alpha:
        print("Since the p-value is less than the significance level (alpha), we reject the null hypothesis.")
        print("Conclusion: The means of the two groups are significantly different.")
    else:
        print("Since the p-value is greater than the significance level (alpha), we fail to reject the null hypothesis.")
        print("Conclusion: The means of the two groups are not significantly different.")


np.random.seed(0)


group1 = np.random.normal(loc=100, scale=10, size=30)
group2 = np.random.normal(loc=110, scale=15, size=30)


independent_t_test(group1, group2)
```

```python
# Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def paired_t_test(before, after, alpha=0.05):

    t_stat, p_value = stats.ttest_rel(before, after)


    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")


    if p_value < alpha:
        print("Since the p-value is less than the significance level (alpha), we reject the null hypothesis.")
        print("Conclusion: There is a significant difference between the two groups.")
    else:
        print("Since the p-value is greater than the significance level (alpha), we fail to reject the null hypothesis.")
        print("Conclusion: There is no significant difference between the two groups.")


    differences = before - after
    plt.figure(figsize=(8, 5))
    plt.hist(differences, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
    plt.axvline(0, color='red', linestyle='--', label='Difference = 0')
    plt.title('Distribution of Differences Between Paired Samples')
    plt.xlabel('Differences (Before - After)')
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()


np.random.seed(0)


before_treatment = np.random.normal(loc=100, scale=10, size=30)
after_treatment = before_treatment + np.random.normal(loc=5, scale=5, size=30)


paired_t_test(before_treatment, after_treatment)
```

```python
#  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence_level=0.95):

    sample_mean = np.mean(sample)
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)


    alpha = 1 - confidence_level
    if sample_size > 30:
        critical_value = stats.norm.ppf(1 - alpha / 2)
    else:
        critical_value = stats.t.ppf(1 - alpha / 2, df=sample_size - 1)


    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))


    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound


np.random.seed(0)


sample_data = np.random.normal(loc=100, scale=15, size=50)
lower, upper = confidence_interval(sample_data, confidence_level=0.95)


print(f"95% Confidence Interval for the sample mean: ({lower:.2f}, {upper:.2f})")
```

```python
# Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence_level=0.95):

    sample_size = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)


    alpha = 1 - confidence_level
    if sample_size > 30:
        critical_value = stats.norm.ppf(1 - alpha / 2)
    else:
        critical_value = stats.t.ppf(1 - alpha / 2, df=sample_size - 1)


    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    return margin_of_error


np.random.seed(0)


sample_data = np.random.normal(loc=100, scale=15, size=50)
moe = margin_of_error(sample_data, confidence_level=0.95)


print(f"Margin of Error for the sample data: {moe:.2f}")
```

```python
#  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.


def bayesian_inference(prior, likelihood, evidence):

    posterior = (likelihood * prior) / evidence
    return posterior


prior = 0.01
likelihood = 0.9
false_positive_rate = 0.05
evidence = (likelihood * prior) + (false_positive_rate * (1 - prior))


posterior = bayesian_inference(prior, likelihood, evidence)

print(f"Posterior probability (probability that the patient has the disease): {posterior:.4f}")
```

```python
# Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np

def calculate_expected_frequencies(observed):

    row_totals = observed.sum(axis=1)


    column_totals = observed.sum(axis=0)


    grand_total = observed.sum()


    expected = np.outer(row_totals, column_totals) / grand_total

    return expected


observed_data = np.array([[30, 10],
                          [20, 40]])


expected_frequencies = calculate_expected_frequencies(observed_data)


print("Expected Frequencies:\n", expected_frequencies)



```

```python
# Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
import scipy.stats as stats


def f_test(sample1, sample2):

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)


    if var1 > var2:
        f_statistic = var1 / var2
    else:
        f_statistic = var2 / var1


    df1 = len(sample1) - 1
    df2 = len(sample2) - 1


    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

    return f_statistic, p_value, var1, var2

np.random.seed(42)
sample1 = np.random.normal(loc=0, scale=2, size=50)
sample2 = np.random.normal(loc=0, scale=3, size=50)


f_statistic, p_value, var1, var2 = f_test(sample1, sample2)


print(f"Variance of Sample 1: {var1:.4f}")
print(f"Variance of Sample 2: {var2:.4f}")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")


if p_value < 0.05:
    print("Reject the null hypothesis: The variances of the two samples are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances of the two samples are not significantly different.")
```

```python
# Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

import numpy as np
import scipy.stats as stats


def anova_test(groups):

    f_statistic, p_value = stats.f_oneway(*groups)

    return f_statistic, p_value


np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=55, scale=5, size=30)
group3 = np.random.normal(loc=60, scale=5, size=30)


groups = [group1, group2, group3]


f_statistic, p_value = anova_test(groups)


print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")


if p_value < 0.05:
    print("Reject the null hypothesis: At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis: The group means are not significantly different.")
```

```python
# Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns


def anova_test(groups):

    f_statistic, p_value = stats.f_oneway(*groups)

    return f_statistic, p_value


np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=55, scale=5, size=30)
group3 = np.random.normal(loc=60, scale=5, size=30)


groups = [group1, group2, group3]


f_statistic, p_value = anova_test(groups)


print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")


if p_value < 0.05:
    print("Reject the null hypothesis: At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis: The group means are not significantly different.")


plt.figure(figsize=(8, 6))
sns.boxplot(data=groups)
plt.title("One-Way ANOVA: Comparison of Group Means", fontsize=16)
plt.xlabel("Groups", fontsize=14)
plt.ylabel("Values", fontsize=14)
plt.xticks([0, 1, 2], ['Group 1', 'Group 2', 'Group 3'])
plt.grid(True)
plt.show()
```

```python
# Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)


factor_a = ['A1', 'A2']

factor_b = ['B1', 'B2', 'B3']


data = []
for a in factor_a:
    for b in factor_b:

        values = np.random.normal(loc=10 + (factor_a.index(a) * 5) + (factor_b.index(b) * 2), scale=2, size=30)
        for value in values:
            data.append([a, b, value])

df = pd.DataFrame(data, columns=['Factor_A', 'Factor_B', 'Value'])


model = ols('Value ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=df).fit()
anova_table = anova_lm(model)


print(anova_table)


```

```python
# Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


def plot_f_distribution(df1, df2, num_points=1000):
    x = np.linspace(0, 5, num_points)
    y = stats.f.pdf(x, df1, df2)


    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
    plt.title(f'F-distribution (df1={df1}, df2={df2})', fontsize=16)
    plt.xlabel('F-value', fontsize=14)
    plt.ylabel('Density', fontsize=14)


    alpha = 0.05
    critical_value = stats.f.ppf(1 - alpha, df1, df2)
    plt.axvline(x=critical_value, color='red', linestyle='--', label=f'Critical F-value (alpha={alpha})')


    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()


plot_f_distribution(df1=5, df2=10)
```

```python
# Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(42)


group1 = np.random.normal(loc=20, scale=5, size=50)


group2 = np.random.normal(loc=25, scale=5, size=50)


group3 = np.random.normal(loc=30, scale=5, size=50)


data = pd.DataFrame({
    'Group': ['Group 1'] * 50 + ['Group 2'] * 50 + ['Group 3'] * 50,
    'Value': np.concatenate([group1, group2, group3])
})


f_stat, p_value = stats.f_oneway(group1, group2, group3)


print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")


plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Value', data=data, palette='Set2')
plt.title('One-Way ANOVA: Comparing Group Means')
plt.xlabel('Group')
plt.ylabel('Value')
plt.show()


if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference between at least two groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the groups.")
```

```python
# Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


np.random.seed(42)


population_mean = 50
population_std = 10
sample_size = 30


sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
t_stat, p_value = stats.ttest_1samp(sample_data, population_mean)


print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"t-statistic: {t_stat}")
print(f"P-value: {p_value}")


alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")


plt.figure(figsize=(8, 6))


plt.hist(sample_data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label=f"Population Mean = {population_mean}")
plt.axvline(sample_mean, color='green', linestyle='dashed', linewidth=2, label=f"Sample Mean = {sample_mean}")


plt.title('Histogram of Sample Data and Hypothesis Testing', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend()

plt.show()
```

```python
# Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
import numpy as np
import scipy.stats as stats


np.random.seed(42)


population_mean = 50
population_variance = 25
sample_size = 30


sample_data = np.random.normal(loc=population_mean, scale=np.sqrt(population_variance), size=sample_size)


sample_variance = np.var(sample_data, ddof=1)
n = len(sample_data)


chi_square_stat = (n - 1) * sample_variance / population_variance


alpha = 0.05
df = n - 1


chi_square_critical_lower = stats.chi2.ppf(alpha / 2, df)
chi_square_critical_upper = stats.chi2.ppf(1 - alpha / 2, df)


print(f"Chi-square Statistic: {chi_square_stat}")
print(f"Critical values for Chi-square (alpha={alpha}):")
print(f"Lower critical value: {chi_square_critical_lower}")
print(f"Upper critical value: {chi_square_critical_upper}")


if chi_square_stat < chi_square_critical_lower or chi_square_stat > chi_square_critical_upper:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized variance.")
```

```python
# Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
import numpy as np
import scipy.stats as stats


successes_1 = 100
sample_size_1 = 250


successes_2 = 120
sample_size_2 = 300

p1 = successes_1 / sample_size_1
p2 = successes_2 / sample_size_2

pooled_proportion = (successes_1 + successes_2) / (sample_size_1 + sample_size_2)


Z = (p1 - p2) / np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1/sample_size_1 + 1/sample_size_2))


p_value = 2 * (1 - stats.norm.cdf(abs(Z)))


print(f"Proportion of successes in Group 1: {p1:.4f}")
print(f"Proportion of successes in Group 2: {p2:.4f}")
print(f"Z-statistic: {Z:.4f}")
print(f"P-value: {p_value:.4f}")


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```
